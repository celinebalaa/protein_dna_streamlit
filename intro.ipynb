<h2><FONT color="palevioletred"> • Qu’est-ce que Streamlit ?    </FONT></h2>
    
<p align="justify"><FONT color="black">  
    <b>Streamlit</b> est une bibliothèque Python open-source permettant la construction d’applications web de manière très simplifiée et reconnue comme étant la <b>plus puissante</b> dans ce domaine. Son point fort est le <u>partage</u>.
    <br> Il est possible de partager facilement des résultats d’expériences dynamiques et modèles dynamiques complexes. 
    <br>Le terme <b>dynamique</b> est ce qui fait toute la différence sur cette bibliothèque. Là ou d’autres ne peuvent montrer que des résultats sous forme "statique" et dévoilent rapidement leurs limites, notamment quand le modèle devient plus complexe nécessitant une entrée conséquente de données par l’utilisateur. Streamlit montre tout son potentiel.</FONT></p>

<h2><FONT color="palevioletred">2) Une première démarche... attention à la marche !</FONT></h2>
<h3><FONT color="pink">Le code génétique</FONT></h3>
  
<p align="justify"><FONT color="black">Au départ nous avons construit le code génétique en utilisant un dictionnaire, ce qui permettait de faire rapidement la liaison entre les codons lus et les acides aminés correspondants.

<h3><FONT color="pink">Lecture de la séquence entrée par l'utilisateur</FONT></h3> 
<p align="justify"><FONThttps://docs.streamlit.io/  color="black">Il n'est pas possible de répondre à la problématique sans lire la séquence entrée sous format FASTA. 
    <br>En effet la première ligne d'un fichier FASTA est ce qu'on appelle une "ligne de description" contenant les informations complémentaires à la séquence (espèces, chromosome...). C'est ce qui fait la spécificité du fichier FASTA.
    <br>C'est pourquoi il est primordiale de récupérer cette ligne dans un champ séparé, afin d'avoir seulement les nucléotides du fragment d'intérêt.
    <br>Nous avons alors récupéré les séquences nucléotidiques sous forme de dictionaire. 
    <br> De plus la commande <b>.strip()</b> nous a permis d'éviter les désagréments liés aux sauts de ligne.</FONT></p>
    


```python
# Récupération de la séquence fournie par l'utilisateur
dna_sequence_input = st.text_area("DNA sequence: ", key="DNA Sequence")

# Fonction permettant de lire et faire un nettoyage des séquences
def read_fasta(dna_sequence_fasta, sequence_type_five):
    dna_sequences = {}
    dna_sequence_fields = dna_sequence_fasta.split("\n")
    dna_id = ""
    for line in dna_sequence_fields:
        if line.startswith(">"):
            dna_id = line[1:].split()[0]
            dna_sequences[dna_id] = ""
        else:
            if sequence_type_five == "RNA":
                line = line.replace("U", "T")
            line = line.upper().strip().replace(" ", "")
            dna_sequences[dna_id] += line
    return dna_sequences
```

<h3><FONT color="pink">Cadres de lecture... pas 1 mais 3 !</FONT></h3> 
<p align="justify"><FONT color="black">    Le cadre de lecture change complètement le résultat. Nous devions donc y accorder une certaine attention. 
    <br>Ainsi à partir du cadre choisi (1, 2 ou 3) par l'utilisateur, celui-ci sera fourni à la fonction comme argument et précisera le nucléotide initialment lu.
    <br>Finalement nous avons ajouté une condition à la boucle quant à la longueur du codon. 
    <br>En effet suivant le cadre de lecture, le codon final pouvait n'être composé d'un, deux ou trois nucléotides. Or trois nucléotides sont nécessaires à la traduction en un acide aminé. Donc on sortira de la boucle si la condition n'est pas respectée.
    <br>Pour chaque codon, nous avons vérifié que les bases qui le forment sont bien des bases nucléotidiques (A,U,T,C,G).
    </FONT></p>  

```python
def dna_protein(dna_seq, reading_frame):
    protein_sequence = ""
    for i in range(reading_frame, len(dna_seq) - reading_frame, 3):
        codon = dna_seq[i:i+3]
        not_base = False
        if len(codon) != 3:
            break
        for base in codon:
            if base not in allowed_bases:
                not_base = True
        if not_base:
            continue
        protein_sequence += genetic_code.get(codon) + ''
    return protein_sequence
```

<h3><FONT color="pink">Affichage et Coloration des acides aminés</FONT></h3> 
<p align="justify"><FONT color="black">Concernant l'affichage des séquences proteïques récupérées sous forme de dictionaire, nous affichons pour chaque séquence son identifiant, récupéré précédemment, et la séquence protéique traduite. 
    <br>En effet comme évoqué précédemment, les acides aminés présentent des caractéristiques bien spécifiques, qui doivent être mises en évidence.
    <br> À présent et grâce à l'utilisation de la bibliothèque Streamlit, nous avons pu mettre en place toutes les fonctionnalités citées précédemment et ajouter la coloration des acides aminés en fonction de leur propriété physico-chimiques.
    <br>En effet après tout le travail effectué en amont, il a fallu colorer les lettres de chaque acide aminé qui présentait un caractère Apolaire / Polaire / charge + / Charge -.
    <br> Pour ce faire, nous avons construit une variable <b>aminoacid</b> dans laquelle nous avons stocké les caractéristiques physico-chimiques correspondantes et associé les acides aminés en conséquence de la manière suivante:</FONT></p>
 
 <p align="justify"><FONT color="black">
    <ul>
        <li>Jaune = Apolaires--> G-A-V-L-I-W-P-M-F</li>
        <li>Vert = Polaires --> S-T-C-N-Q-Y</li>
        <li>Rouge = Chargés + --> K-R-H </li>
        <li>Bleu = Chargés - --> D-E</li>
   </ul>
</FONT></p>
